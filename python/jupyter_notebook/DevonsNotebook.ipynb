```python
# Use a function with a tuple as a response

def my_attack(player_health, weapon_dmg, attack_power):
    # player remaining health, then damage done as a tuple output
    damage_done = weapon_dmg + attack_power
    if damage_done > player_health:
        player_health = 0
    else:
        player_health = player_health - damage_done

    return player_health, damage_done

result_tuple = my_attack(100, 17.5, 12)

print("")
print(f"Result of attack: {result_tuple}")
print("")
```

```python
# Use a function to get player current health from key/value pair
# then execute attack function.
player_1 = {
    "name": "JohnnyReb",
    "health": 212,
    "weapon_dmg": 24,
    "attack_power": 10,
}

player_2 = {
    "name": "Toughskin",
    "health": 250,
    "weapon_dmg": 75,
    "attack_power": 30,
}

def attack_player(player_to_attack, attacking_player):
    # returns attacked_player_stats, attack_amount
    
    # Obtain attack_player function variables
    original_health = player_to_attack['health']
    player_health = player_to_attack["health"]
    attacker_weapon_dmg = attacking_player["weapon_dmg"]
    attacker_attack_power = attacking_player["attack_power"]

    # Get attack damage and apply to victim
    attack_damage = attacker_weapon_dmg + attacker_attack_power

    if player_health < attack_damage:
        player_health = 0
    else:
        player_health -= attack_damage

    player_to_attack["health"] = player_health
    player_attack_damage = attack_damage

    return original_health, player_to_attack, player_attack_damage

# perform attack on player_1
original_health, player_to_attack, attacker_damage_done = attack_player(player_1, player_2)

print("")
print(f"Attacked Player Health:")
print(f"    {player_to_attack['name']}")
print(f"        Original Health: {original_health}")
print(f"        Damage Taken => {attacker_damage_done}")
print(f"        Remaining Health => {player_to_attack['health']}")

print("")
print("--- AFTER ATTACK REPORT ---")
print("")

print(f"{player_1['name']} Stats >>")
print(f"    Health: {player_1['health']}")
print(f"    Weapon Damage: {player_1['weapon_dmg']}")
print(f"    Attack Power: {player_1['attack_power']}")
print("")

print(f"{player_2['name']} Stats >>")
print(f"    Health: {player_2['health']}")
print(f"    Weapon Damage: {player_2['weapon_dmg']}")
print(f"    Attack Power: {player_2['attack_power']}")
print("")
```
